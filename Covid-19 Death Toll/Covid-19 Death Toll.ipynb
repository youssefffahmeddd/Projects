## Covid-19 Death Toll

In this project I'm going to work with a supervised dataset from the world health organization 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
top20 = pd.read_csv(r"C:\Users\youss\Desktop\top20_deathtoll.csv")
death_toll = pd.read_csv(r"C:\Users\youss\Desktop\covid_avg_deaths.csv")
print(top20)
```

```python
print(death_toll)
```

**I am going to use matplotlib's OO interface to visualize the data (The OO inteface is usually used to visualize the explainatory visualization as it is more flexible and gives more power)**

```python
fig, ax = plt.subplots()
ax.barh(top20['Country_Other'], top20['Total_Deaths'], color='#bc131f', height=0.45)
for location in ['left','bottom','top','right']:
    ax.spines[location].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(left=False, top=False)
ax.set_xticks([0, 150000, 300000])
ax.set_xticklabels(['0', '150,000', '300,000'], color='grey')
ax.text(-60000,25, 'Top 20 Covid-19 Deaths By Country', weight='bold', size=20)
ax.text(-60000, 23, 'The data was collected by the World Health Organization in 2020')
ax.axvline(150000, c='#666666', ymin=0.045, alpha=0.5)
plt.show()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1)
axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(death_toll['Month'], death_toll['New_deaths'], alpha=0.5, color='#bc131f')
    for location in ['top', 'bottom', 'left', 'right']:
        ax.spines[location].set_visible(False)
        ax.set_xticklabels([])
        ax.tick_params(left=False, bottom=False)
        ax.set_yticklabels([])
ax1.text(0.5, -80, '0', alpha=0.5)
ax1.text(3.5, 2000, '1,844', alpha=0.5)
ax1.text(7.5, 1300, '1,020', alpha=0.5)
ax1.text(11.5, 2400, '2,247', alpha=0.5)
ax1.plot(death_toll['Month'][0:3], death_toll['New_deaths'][0:3], color='#bc131f', linewidth=3)
ax2.plot(death_toll['Month'][2:6], death_toll['New_deaths'][2:6], color='#bc131f', linewidth=3)
ax3.plot(death_toll['Month'][5:10], death_toll['New_deaths'][5:10], color='#bc131f', linewidth=3)
ax4.plot(death_toll['Month'][9:12], death_toll['New_deaths'][9:12], color='#bc131f', linewidth=3)
ax1.text(0.5, 3600, 'The virus kills 851 people each day', size=14, weight='bold')
ax1.text(0.5, 3150, 'Average number of daily deaths per month in the US', size=12)
ax1.text(1.1, -500, 'Jan - Mar', color='#af0b1e', weight='bold', rotation=3)
ax2.text(3.7, 700, 'Mar - Jun', color='#af0b1e', weight='bold')
ax3.text(7.1, 400, 'Jun - Oct', color='#af0b1e', weight='bold')
ax4.text(10.7, 500, 'Oct - Dec', color='#af0b1e', weight='bold', rotation=35)
plt.show()
```
